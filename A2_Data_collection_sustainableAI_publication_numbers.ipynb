## Sustainability Patents

```python
# Read the Excel file into a DataFrame
# WIPO
WIPO1 = pd.read_csv('Sustain_WIPO_IPC1_mit_auth.csv', delimiter=';')
WIPO2 = pd.read_csv('Sustain_WIPO_IPC2_mit_auth.csv', delimiter=';')
WIPO3 = pd.read_csv('Sustain_WIPO_IPC3_mit_auth.csv', delimiter=';')
WIPO4 = pd.read_csv('Sustain_WIPO_IPC4_mit_auth.csv', delimiter=';')
WIPO5 = pd.read_csv('Sustain_WIPO_IPC5_mit_auth.csv', delimiter=';')

# ENVTECH
# IPC
ENVTECH1 = pd.read_csv('Sustain_ENVTECH_IPC1_mit_auth.csv', delimiter=';')
ENVTECH2 = pd.read_csv('Sustain_ENVTECH_IPC2_mit_auth.csv', delimiter=';')
ENVTECH3 = pd.read_csv('Sustain_ENVTECH_IPC3_mit_auth.csv', delimiter=';')
ENVTECH4 = pd.read_csv('Sustain_ENVTECH_IPC4_mit_auth.csv', delimiter=';')
ENVTECH5 = pd.read_csv('Sustain_ENVTECH_IPC5_mit_auth.csv', delimiter=';')
ENVTECH5 = pd.read_csv('Sustain_ENVTECH_IPC5_mit_auth.csv', delimiter=';')

# CPC A
ENVTECH6 = pd.read_csv('envtech_cpc_mit_auth.csv', delimiter=';')
ENVTECH7 = pd.read_csv('envtech_cpc2_mit_auth.csv', delimiter=';')
ENVTECH8 = pd.read_csv('envtech_cpc3_mit_auth.csv', delimiter=';')
ENVTECH9 = pd.read_csv('envtech_cpc4_mit_auth.csv', delimiter=';')
ENVTECH10 = pd.read_csv('envtech_cpc5_mit_auth.csv', delimiter=';')
ENVTECH11 = pd.read_csv('envtech_cpc6_mit_auth.csv', delimiter=';')
ENVTECH12 = pd.read_csv('envtech_cpc7_mit_auth.csv', delimiter=';')

# CPC B
ENVTECH13 = pd.read_csv('Sustain_ENVTECH_CPC1_mit_auth.csv', delimiter=';')
ENVTECH14 = pd.read_csv('Sustain_ENVTECH_CPC2_mit_auth.csv', delimiter=';')
ENVTECH15 = pd.read_csv('Sustain_ENVTECH_CPC3_mit_auth.csv', delimiter=';')
ENVTECH16 = pd.read_csv('Sustain_ENVTECH_CPC4_mit_auth.csv', delimiter=';')
ENVTECH17 = pd.read_csv('Sustain_ENVTECH_CPC5_mit_auth.csv', delimiter=';')
ENVTECH18 = pd.read_csv('Sustain_ENVTECH_CPC6_mit_auth.csv', delimiter=';')

# Y-Codes
YCODE1 = pd.read_csv('y02EPtop.csv', delimiter=';')
YCODE2 = pd.read_csv('y02EPbottom.csv', delimiter=';')
YCODE3 = pd.read_csv('Sustain_YCODES_CPC1_mit_auth.csv', delimiter=';')
YCODE4 = pd.read_csv('Sustain_YCODES_CPC2_mit_auth.csv', delimiter=';')
```

```python
WIPO1
```

```python
Sustain_WIPO_complete = pd.concat([WIPO1, WIPO2, WIPO3, WIPO4, WIPO5]).reset_index(drop=True)  # drop=True removes old indices
Sustain_ENVTECH_IPC_complete = pd.concat([ENVTECH1, ENVTECH2, ENVTECH3, ENVTECH4, ENVTECH5]).reset_index(drop=True)
Sustain_ENVTECH_CPC_complete = pd.concat([ENVTECH13, ENVTECH14, ENVTECH15, ENVTECH16, ENVTECH17, ENVTECH18]).reset_index(drop=True)
Sustain_YCODES_complete = pd.concat([YCODE3, YCODE4]).reset_index(drop=True)
```

```python
Sustain_WIPO_complete_deduped = Sustain_WIPO_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
               .agg({"ipc_class_symbol": lambda x: list(set(x))}) \
               .reset_index()
Sustain_ENVTECH_IPC_complete_deduped = Sustain_ENVTECH_IPC_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
                .agg({"ipc_class_symbol": lambda x: list(set(x))}) \
                .reset_index()
Sustain_ENVTECH_CPC_complete_deduped = Sustain_ENVTECH_CPC_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
                .agg({"cpc_class_symbol": lambda x: list(set(x))}) \
                .reset_index()
Sustain_YCODES_complete_deduped = Sustain_YCODES_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
                .agg({"cpc_class_symbol": lambda x: list(set(x))}) \
                .reset_index()
```

```python
print(len(Sustain_WIPO_complete), len(Sustain_WIPO_complete_deduped))
print(len(Sustain_ENVTECH_IPC_complete), len(Sustain_ENVTECH_IPC_complete_deduped))
print(len(Sustain_ENVTECH_CPC_complete), len(Sustain_ENVTECH_CPC_complete_deduped))
print(len(Sustain_YCODES_complete), len(Sustain_YCODES_complete_deduped))
```

```python
Sustain_ENVTECH_complete = pd.concat([Sustain_ENVTECH_IPC_complete_deduped, Sustain_ENVTECH_CPC_complete_deduped], sort=True).reset_index(drop=True)
```

```python
Sustain_ENVTECH_complete["ipc_class_symbol"] = Sustain_ENVTECH_complete["ipc_class_symbol"].apply(lambda x: x if isinstance(x, list) else ([x] if pd.notna(x) else []))
Sustain_ENVTECH_complete["cpc_class_symbol"] = Sustain_ENVTECH_complete["cpc_class_symbol"].apply(lambda x: x if isinstance(x, list) else ([x] if pd.notna(x) else []))



Sustain_ENVTECH_deduped = Sustain_ENVTECH_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
               .agg({
        "ipc_class_symbol": lambda x: list(set([item for sublist in x for item in sublist])),  # Flatten & unique
        "cpc_class_symbol": lambda x: list(set([item for sublist in x for item in sublist]))   # Flatten & unique
    }) \
    .reset_index()
```

```python
print(len(Sustain_ENVTECH_complete), len(Sustain_ENVTECH_deduped))
```

```python
# Find the list with the most entries
max_ipc_list = max(Sustain_ENVTECH_complete['ipc_class_symbol'], key=len)

# Print the list with the most entries
print(max_ipc_list)
# Find the index of the row with the list with the most entries
max_ipc_row_index = Sustain_ENVTECH_complete['ipc_class_symbol'].apply(len).idxmax()

# Print the row with the list with the most entries
print(Sustain_ENVTECH_complete.loc[max_ipc_row_index])
```

```python
Sustain_complete = pd.concat([Sustain_WIPO_complete_deduped, Sustain_ENVTECH_deduped, Sustain_YCODES_complete_deduped], sort=True).reset_index(drop=True)
```

```python
Sustain_complete["ipc_class_symbol"] = Sustain_complete["ipc_class_symbol"].apply(lambda x: x if isinstance(x, list) else ([x] if pd.notna(x) else []))
Sustain_complete["cpc_class_symbol"] = Sustain_complete["cpc_class_symbol"].apply(lambda x: x if isinstance(x, list) else ([x] if pd.notna(x) else []))



Sustain_complete_deduped = Sustain_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
               .agg({
        "ipc_class_symbol": lambda x: list(set([item for sublist in x for item in sublist])),  # Flatten & unique
        "cpc_class_symbol": lambda x: list(set([item for sublist in x for item in sublist]))   # Flatten & unique
    }) \
    .reset_index()
```

## Artificial Intelligence

```python
# combinations
AI_comb1 = pd.read_csv('AI_combinations1_mit_auth.csv', delimiter=';')
AI_comb2 = pd.read_csv('AI_combinations2_mit_auth.csv', delimiter=';')
AI_comb3 = pd.read_csv('AI_combinations3_mit_auth.csv', delimiter=';')
AI_comb4 = pd.read_csv('AI_combinations4_mit_auth.csv', delimiter=';')
AI_comb5 = pd.read_csv('AI_combinations5_mit_auth.csv', delimiter=';')
AI_comb6 = pd.read_csv('AI_combinations6_mit_auth.csv', delimiter=';')
AI_comb7 = pd.read_csv('AI_combinations7_mit_auth.csv', delimiter=';')
AI_comb8 = pd.read_csv('AI_combinations8_mit_auth.csv', delimiter=';')
AI_comb9 = pd.read_csv('AI_combinations9_mit_auth.csv', delimiter=';')
AI_comb10 = pd.read_csv('AI_combinations10_mit_auth.csv', delimiter=';')

# CPC & IPC
AI_CPC = pd.read_csv('AI_CPC_mit_auth.csv', delimiter=';')
AI_IPC = pd.read_csv('AI_IPC_mit_auth.csv', delimiter=';')

# keywords
AI_keyword1 = pd.read_csv('AI_keyword1_mit_auth.csv', delimiter=';')
AI_keyword2 = pd.read_csv('AI_keyword2_mit_auth.csv', delimiter=';')
AI_keyword3 = pd.read_csv('AI_keyword3_mit_auth.csv', delimiter=';')
AI_keyword4 = pd.read_csv('AI_keyword4_mit_auth.csv', delimiter=';')
AI_keyword5 = pd.read_csv('AI_keyword5_mit_auth.csv', delimiter=';')
AI_keyword6 = pd.read_csv('AI_keyword6_mit_auth.csv', delimiter=';')
AI_keyword7 = pd.read_csv('AI_keyword7_mit_auth.csv', delimiter=';')
AI_keyword8 = pd.read_csv('AI_keyword8_mit_auth.csv', delimiter=';')
```

```python
AI_complete = pd.concat([AI_comb1, AI_comb2, AI_comb3, AI_comb4, AI_comb5, AI_comb6, AI_comb7, AI_comb8, AI_comb9, AI_comb10, AI_CPC, AI_IPC, AI_keyword1, AI_keyword2, AI_keyword3, AI_keyword4, AI_keyword5, AI_keyword6, AI_keyword7, AI_keyword8], sort=True).reset_index(drop=True)
```

```python
AI_complete_deduped = AI_complete.groupby(["PUBLN_NR", "PUBLN_KIND", "publn_auth", "publn_date"]) \
                .agg({  "ipc_class_symbol": lambda x: list(set(x.dropna())),
                        "cpc_class_symbol": lambda x: list(set(x.dropna()))}) \
                .reset_index()
```
